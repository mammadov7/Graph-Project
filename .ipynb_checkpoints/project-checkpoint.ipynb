```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
def prob( G ):
    degrees = np.array(G.degree,dtype='float' )
    degrees[:,0] += 1
    degrees[:,1] = degrees[:,1] / np.sum(degrees[:,1])
    return degrees

def cum_prob( G, delta  = 0):
    degrees = np.array(G.degree,dtype='float' )[:-1,1]
    probs = (degrees + delta) / np.sum(degrees+delta)
    #Calculation de probabilite cumulative pour tirage
    #Ex: [0.2 0.3 0.5] -> [0.2 0.5 1.0] 
    for i in range(1,len(probs)):
        probs[i] += probs[i-1] 
    return probs

def tire_sommet(G,delta = 0):
    cum_probs = cum_prob(G,delta)
    _random = np.random.random()
    for i in range(len(cum_probs)):
        if _random <= cum_probs[i]:
            return i
        
def Barabasi(n,delta = 0,directed=0):
    if directed:
        G = nx.DiGraph()
    else :
        G = nx.Graph()
    G.add_node(0)
    G.add_edge(0,0)
    for i in range(1,n):
        G.add_node(i)
        G.add_edge(tire_sommet(G,delta),i)
    return G

def matrice_adjacence(G):
    n = len(G.nodes)
    matrice = np.zeros([n,n])
    for i in range(n):
        matrice[i, list(G.adj[i]) ] += 1
    return matrice
```

```python
# nx.draw_networkx(g)
def reg_lin(G):
    X = np.log(prob(G))    
    Y = np.copy(X[:,1]) 
    X[:,1] = np.ones_like(X[:,1])
    XTX = np.dot(X.T,X)
    XTY = np.dot(X.T,Y)
    a, b = np.dot( np.linalg.inv(XTX), XTY)
    alpha = a
    c = (np.e)**b
    return alpha,c

def draw_random_graph(g):
    N = len(g.nodes)
    X = np.zeros( N )
    Y = np.zeros( N )
    m = matrice_adjacence(g)
    for i in range( N ):
        X[i] = np.random.random()
        Y[i] = np.random.random()
    for i in range( N ):
        for j in range( i,N ):
            if m[i][j] :
                plt.plot([X[i], X[j]],[Y[i], Y[j]], 'red')
    plt.plot(X, Y,'o')
    plt.show()
    
def bfs(graph):
    N = len(graph.nodes)
    dist = np.zeros([N,N])
    graph.remove_edge(0,0)
    for n in range(N):
        # tous les sommets visités
        explored = []
        # sommets à vérifier
        queue = [n]
        # continuez à boucler jusqu'à ce qu'il reste des sommets à vérifier
        while queue:
            # pop (premier sommet) de la file d'attente
            node = queue.pop(0)
            if node not in explored:
                # ajouter le sommet à la liste des nœuds vérifiés
                explored.append(node)
                neighbours = list(graph.adj[node])    
                for i in neighbours:
                    if i not in explored:
                        dist[n][i] += (1+dist[n][node])
                # ajouter des voisins du sommet à la file d'attente
                    queue.append(i)
    graph.add_edge(0,0)
    return dist

def true_points(G, d=2, lr = 0.01, tol = 0.01):
    N = len(G.nodes)
    Distance = bfs(G)
    Dmax = np.max(Distance)
    D = lambda i,j: Distance[i][j]/Dmax
    M = np.zeros( [N,d] )
    for i in range( N ):
        for j in range(d):
            M[i][j] = np.random.random()
    p_E = 10**10 
    E = p_E - 1  
    # continuez à boucler jusqu'à ce qu'il difference des energies est moins de tolerence
    while( (p_E - E) > tol ):
        p_E = E # Energie precedente
        E = 0 # Energie
        for i in range(N): # pour chaque i < j
            for j in range(i+1,N):
                r = (np.linalg.norm( M[i] - M[j]) - D(i,j) * np.sqrt(2)) * (  M[i] - M[j] ) 
                r /= ( (D(i,j)**2) * np.linalg.norm( M[i] - M[j]) )  
                M[i] = M[i] - lr*r # k = i, dE > 0 
                M[j] = M[j] + lr*r # k = j, dE < 0
                # Mise a jour de l'energie E 
                E += ((np.linalg.norm( M[i] - M[j])/np.sqrt(2) - D(i,j))**2 / (D(i,j)**2)) 
    return M

def draw_optimal_graph(G, lr = 0.001, tol = 0.01):
    N = len(G.nodes)
    M = true_points(G,2,lr,tol)
    m = matrice_adjacence(G)
    for i in range( N ):
        for j in range( i, N ):
            if m[i][j] :
                plt.plot([M[i][0], M[j][0]],[M[i][1], M[j][1]], 'red')
    plt.plot(M[:,0], M[:,1],'o')
    
def draw_3d(G, lr = 0.01, tol = 0.01):
    N = len(G.nodes)
    ax = plt.axes(projection='3d')
    M = true_points(G,d=3,lr=lr,tol=tol)
    m = matrice_adjacence(G)
    for i in range( N ):
        for j in range( i, N ):
            if m[i][j] :
                ax.plot3D([M[i][0], M[j][0]],[M[i][1], M[j][1]],[M[i][2], M[j][2]] , 'red')
    ax.scatter3D(M[:,0], M[:,1], M[:,2], c=M[:,2], cmap='viridis')
```

```python
def Spectral_Clustering( G, M, k ):
    D = np.array(G.degree,dtype='float' )[:,1]
    A = matrice_adjacence(G)
    # Laplacian matrix 
    L = np.diag(D)-A
    eigval, eigvec = np.linalg.eig(L)
    eigvec = np.real( eigvec )
    eigval = np.real( eigval )
    kfirst = np.argsort(eigval)[:k]
    Uk = eigvec[:,kfirst]
    c,S = K_means(Uk,k)

    plt.figure(figsize=(7,4))
    for j in range(k):
        x1 = [M[i,0] for i in range(M.shape[0]) if S[i,j]==1]
        x2 = [M[i,1] for i in range(M.shape[0]) if S[i,j]==1]
        plt.scatter(x1,x2, c=clusters_col[j])
        c[j][0] = np.mean(x1)  
        c[j][1] = np.mean(x2)  
    plt.scatter(c[:,0].reshape(-1), c[:,1].reshape(-1), c='black', marker='x')
    plt.title('Spectral Clustering')
    plt.show()
    
def K_means(X, K):
    """
    Returns:
        c (np.array): matrix containing the cluster centeres means (K,d)
        S (np.array): binary matrix containing 1-hot encodings of the cluster membership (N,K)
    """
    E=10**(-8) # Tolerance
    N = X.shape[0]
    #initialion
    initial_indices = np.random.choice(np.arange(N), size=K, replace=False)
    c = X[initial_indices, :]
    c_temp = 100
    while(np.abs(np.mean(c-c_temp))>E):
        dists = np.zeros((N,K))
        for k in range(K):
            dists[:,k] = np.linalg.norm(X-c[k,:], axis=1)      
        #step(1)
        cluster_membership = np.argmin(dists, axis=1)
        S = np.zeros((N,K))
        for i in range(N):
            S[i,cluster_membership[i]] = 1
        #step(2)
        c_temp = c
        c = S.T@X/ np.sum(S,0).reshape(-1,1)
    return c, S

def plot_Kmeans(M, k):
    c,S = K_means(M,k)
    clusters_col = ["red", "blue", "yellow", "purple", "cyan"] 
    plt.figure(figsize=(7,4))
    for j in range(k):
        x1 = [M[i,0] for i in range(M.shape[0]) if S[i,j]==1]
        x2 = [M[i,1] for i in range(M.shape[0]) if S[i,j]==1]
        plt.scatter(x1,x2, c=clusters_col[j])
    plt.scatter(c[:,0].reshape(-1), c[:,1].reshape(-1), c='black', marker='x')
    plt.title('K-means')
    plt.show() 
```

```python
def SBM(N, K, p=.8, q=.05):
    G = nx.Graph()
    G.add_nodes_from(range(N))
    for i in range(N):
        for j in range(i + 1, N):
            theta = p if (i % K == j % K) else q
            if np.random.random() < theta:
                G.add_edge(i, j)
    return G
t = 200
G = SBM(200,5)
pos = nx.spring_layout(G)
M = np.zeros([t,2])
for i in range(t):
    M[i][0] = pos[i][0]
    M[i][1] = pos[i][1]
```

```python
def power(A, N_max, TOL):
    n = A.shape[0]
    V = np.random.rand(n) # initialize to random vector
    V /= V.sum() # make v sum to 1
    AV = np.dot(A,V)
    err = np.linalg.norm( AV - V )
    for _ in range(N_max):
        if( err < TOL ) : break
        V = AV
        AV = np.dot(A,V)
        err = np.linalg.norm( AV - V )
    return V
def power(A, N_max, TOL):
    n = A.shape[0]
    V = np.random.rand(n) # initialize to random vector
    V /= V.sum() # make v sum to 1
    err = 1
    for _ in range(N_max):
        if( err < TOL ) : break
        prev_V = V
        V = np.dot(A,V)
        err = np.linalg.norm( V - prev_V )
    return V
```

```python
def PageRank( G, e = 0.15, tol = 1e-6 ):
    n = len(G.nodes)
    M = matrice_adjacence(G)
    M[0][0] = 0
    for i in range(n):
        tmp = np.sum(M[i,:])
        if tmp > 0:
            M[i] /= tmp
        else :
            M[i] =  np.ones(n)/n
    P = (1-e)*M.T + (e/n)*np.ones_like(M)
    return power(P, N_max=1000, TOL=tol),P
```
