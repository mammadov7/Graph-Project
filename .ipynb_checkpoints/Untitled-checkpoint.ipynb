```python
import numpy as np
```

```python
ratings = []
with open("/home/ali/Desktop/INF583/tp1/ratings.csv") as f:
    i = 0
    for line in f:
        if not i: 
            i += 1
            continue
        token = line.split(',')
        userId = int(token[0])
        movieId = int(token[1])
        rating = float(token[2])
        timestamp = str(token[3])
        ratings.append([userId, movieId, rating, timestamp])
```

```python
for i in range(len(ratings)):
    ratings[i][3] = list(ratings[i][3])
```
